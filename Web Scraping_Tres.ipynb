```python
rows=[]
url_prefix='https://www.tresfashion.co/tres_product_index/all/tres/sale'
for e in a_tags:
    img=e.find('img')
    name=e.find('h2')
    price=e.find('h4')
    sale_price=e.find('h3')
    src=img['src']
    img_html=f'<img src="{src}" />'
    link=e['href']
    href_html=f'<a href="{link}">{name}</a>'
    rows.append((name,price,sale_price, link, src, href_html, img_html))
```

```python
rows
```

```python
df=pd.DataFrame(rows, columns=['name','price','sale price', 'link', 'img_url', 'href_html', 'img_html'])
```

## <div style="background-color:pink;color:white;padding:10px;"><span style="color:white"><b>Clean Data</b></span></div>


```python
tres['name']=tres['name'].str.replace(r'004', 'IV', regex = False)
tres['name']=tres['name'].str.replace(r'02', 'II', regex = False)
tres['name']=tres['name'].str.replace(r'01', 'I', regex = False)
tres['name']
```

```python
dll=tres['name'].str.extract(r'(\D{2}\d\D)(\D+)(\D{3}\d\D)')
dll.columns=['nan1','name',	'nan2']
dll['name']
tres['name']=dll['name']
tres['name']
```

```python
tres['price']=tres['price'].str.replace(r',', '', regex = False)
tres['price']
```

```python
tres.info()
```

```python
dll=tres['price'].str.extract(r'(\s\d+)').astype(int)
dll.columns=['price']
dll['price']
tres['price']=dll['price']
tres['price']
```

```python
tres.info()
```

```python
tres['sale price']=tres['sale price'].str.replace(r',', '', regex = False)
tres['sale price']
```

```python
dll=tres['sale price'].str.extract(r'(\s\d+)').astype(int)
dll.columns=['sale price']
dll['sale price']
tres['sale price']=dll['sale price']
tres['sale price']
```

```python
tres.set_index('name', inplace=True)
tres.head()
```

```python
tres.info()
```
